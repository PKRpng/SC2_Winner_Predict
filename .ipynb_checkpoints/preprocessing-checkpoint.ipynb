```python
import sc2reader
import pandas as pd
import time
import re
import os
```

```python
path = 'SC2_raw_data/Complete_Replays/'

files = []

winner = []
map_name = []
player_one_race = []
player_two_race = []
game_length = []
is_best_5 = []
player_one = []
player_two = []
date = []
```

```python
def collect_files(path):
    '''append replays to a list from all replay folders'''
    
    file_list = []
    
    for root, directories, files in os.walk(path):
        for name in files:
            file_list.append(os.path.join(root, name))
    
    return(file_list)
```

```python
def collect_data(files):
    '''collect game data'''
    display_interval = 500
    count = 0
    
    for sc2_replay in files:
        datetime = ''
        game_winner = ''
        player1 = ''
        player2 = ''
        race1 = ''
        race2 = ''
        glength = ''
        mapn = ''
        
        replay = sc2reader.load_replay(sc2_replay, load_map=False)
        
        if len(replay.teams) < 3 and len(replay.teams) > 0 and replay.winner != None:
            try:
                #input data into temporary vars
                datetime = str(replay.date)[:10]
                game_winner = str(replay.winner).split()[5]
                player1 = str(replay.teams[0]).split()[5]
                player2 = str(replay.teams[1]).split()[5]
                race_one = str(replay.teams[0]).split()[6]
                race_two = str(replay.teams[0]).split()[6]
                glength = str(replay.game_length)
                mapn = replay.map_name
                
                race_one = re.sub('[^A-Za-z]+', '', race_one)
                race_two = re.sub('[^A-Za-z]+', '', race_two)
                
                #if no exceptions are raised then append to global list
                date.append(datetime)

                winner_name.append(game_winner)

                player_one.append(player1)
                player_two.append(player2)

                player_one_race.append(race_one)
                player_two_race.append(race_two)

                map_name.append(mapn)

                game_length.append(glength)
                
                if game_winner == player1:
                    p_one_winner.append(1)
                else: p_one_winner.append(0)
                
            except Exception as e:
                print(e)
            
        if count % display_interval == 0:
            print(len(files)-count, 'left to parse.')
        
        count += 1
            
    return()
```

```python
files = collect_files(path)

collect_data(files)
```

```python
df = pd.DataFrame({'date':date,
                    'map_name':map_name, 
                    'player_one_race':player_one_race,
                    'player_two_race':player_two_race,
                    'player_one':player_one, 
                    'player_two':player_two, 
                    'game_length':game_length, 
                    'winner_name':winner_name,
                    'p_one_winner':p_one_winner})
```
